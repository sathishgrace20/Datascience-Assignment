```python
#Q8:Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.(Make decision using Hypothesis Testing)
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

d = dataset[(dataset['degree_t'] == 'Sci&Tech') | (dataset['specialisation'] == 'Mkt&HR')]

model = ols("salary ~ C(degree_t) + C(specialisation) + C(degree_t):C(specialisation)", data=d).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
```

```python
anova_table
```

```python
#pvalue is 0.002964293
#Null Hypothesis --> There is no similarity between the mk&hr and sci&tech
#Alternate hypothesis -->There is similarity between the mk&hr and sci&tech
#Reject null hypo if the pvalues is <=0.05 
#0.029
#We can reject the null hypo, we accept alternate hypo
```

```python
#Q9:Convert the normal distribution to standard normal distribution for salary column
def stdNBgraph(dataset):
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    values=[i for i in dataset]
    z_score=[((j-mean)/std) for j in values]
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
```

```python
stdNBgraph(dataset['salary'])
```

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax=sns.distplot(dataset,kde=True, kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    #Generate sample
    sample = dataset
    #Calculate parameters
    sample_mean=sample.mean()
    sample_std=sample.std()
    print('Mean=%.3f, standard Deviation=%.3f' %(sample_mean, sample_std))
    #define the distribution
    dist= norm(sample_mean,sample_std)
    #Sample probability for range of outcomes
    values=[value for value in range(startrange,endrange)]
    probabilities=[dist.pdf(value) for value in values]
    prob=sum(probabilities)
    print("The area between range({},{}:{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset['salary'],700000,900000)
```
